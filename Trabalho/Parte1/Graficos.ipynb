### Gráficos

```python
import pandas as pd
import matplotlib.pyplot as plt


# Carregar o dataset tratado
url = "http://dados.recife.pe.gov.br/dataset/e2bd8f0b-1d62-4323-8159-8ebd6ed9eb4a/resource/77c885c4-76ca-45eb-9209-06c5d217122d/download/licenciamento_urbanistico.csv"

data = pd.read_csv(url, sep=';')

```

```python
# Remover linhas com valores não numéricos nas colunas de interesse
data = data[pd.to_numeric(data['valor_taxa'], errors='coerce').notnull()]
data = data[pd.to_numeric(data['valor_pago'], errors='coerce').notnull()]

```

```python
# Converter colunas para tipo numérico
data['valor_taxa'] = pd.to_numeric(data['valor_taxa'])
data['valor_pago'] = pd.to_numeric(data['valor_pago'])
```

```python
# Gráfico de dispersão das variáveis numéricas
plt.figure(figsize=(10, 8))
plt.scatter(data['valor_taxa'], data['valor_pago'], alpha=0.5)
plt.title('Dispersão entre Valor da Taxa e Valor Pago')
plt.xlabel('Valor da Taxa')
plt.ylabel('Valor Pago')
plt.grid(True)
plt.show()
```

### O gráfico de dispersão mostra a relação entre duas variáveis numéricas: o eixo x representa o 'Valor da Taxa' e o eixo y representa o 'Valor Pago'. Cada ponto no gráfico representa um processo de licenciamento urbanístico, mostrando quanto foi pago em relação à taxa estabelecida. O padrão dos pontos revela se há uma relação entre os valores pagos e as taxas, ajudando a identificar tendências nos dados.

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "http://dados.recife.pe.gov.br/dataset/e2bd8f0b-1d62-4323-8159-8ebd6ed9eb4a/resource/77c885c4-76ca-45eb-9209-06c5d217122d/download/licenciamento_urbanistico.csv"

data = pd.read_csv(url, sep=';')


# Contar a ocorrência de cada bairro e ordenar em ordem decrescente
contagem_bairros = data['bairro'].value_counts()

# Selecionar os 10 bairros com mais ocorrências
top_10_bairros = contagem_bairros.head(10)

# Criar o gráfico de barras
plt.figure(figsize=(10, 6))
top_10_bairros.plot(kind='bar', color='skyblue')
plt.title('Top 10 Bairros com Mais Ocorrências de Licenciamento Urbanístico')
plt.xlabel('Bairro')
plt.ylabel('Quantidade de Ocorrências')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
