```python
import pandas as pd

# Carregue o conjunto de dados
df = pd.read_csv("/content/77c885c4-76ca-45eb-9209-06c5d217122d.csv")

# Verifique as colunas disponíveis
print(df.columns)

# Selecione a variável numérica para análise
variavel_numerica = 'areatotalconstruida'

# Verificar se a coluna está presente no DataFrame
if variavel_numerica in df.columns:
    # Calcule estatísticas descritivas básicas
    estatisticas_descritivas = df[variavel_numerica].describe()

    # Identifique valores discrepantes usando a regra do IQR
    Q1 = df[variavel_numerica].quantile(0.25)
    Q3 = df[variavel_numerica].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    valores_discrepantes = df[(df[variavel_numerica] < limite_inferior) | (df[variavel_numerica] > limite_superior)]

    # Exiba estatísticas descritivas e valores discrepantes
    print("Estatísticas Descritivas:")
    print(estatisticas_descritivas)
    print("\nValores Discrepantes:")
    print(valores_discrepantes)

```
