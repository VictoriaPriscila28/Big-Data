Conclusões: Percebe-se que no primeiro gráfico, referente aos dados de acidentes aéreos ocorridos no Brasil no período de 2014 até o 1° semestre de 2024, há um aumento gradativo ao longo dos anos, atingindo seu ápice em 2023. Infere-se que em 2024 os números serão ainda maiores, uma vez que já alcançaram 839 casos e o ano ainda não terminou. Além disso, é possível observar que no período de 2020 a 2022 os casos foram semelhantes, coincidindo com a crise sanitária do COVID-19 e o consequente lockdown, impactando negativamente o setor do turismo, o que levanta questionamentos sobre a ausência de uma diminuição significativa nos números. Destaca-se que 2023 registrou o maior número de casos, com um aumento de aproximadamente 125% em relação ao ano anterior. Quais seriam as possíveis causas para tantos acidentes?

Quanto ao gráfico dos acidentes em Pernambuco em comparação com o país, também se observa um aumento durante a pandemia, com um incremento de 50% no primeiro semestre em relação ao anterior. Isso sugere a necessidade de um estudo comparativo da média nacional com outros estados.
